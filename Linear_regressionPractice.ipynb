```python
import pandas as pd
```

```python
salary = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
salary.head()
```

```python
salary.columns
```

```python
y = salary['Salary']
y
```

```python
salary.info()
```

```python
y.value_counts()
```

```python
X = salary[['Experience Years']]
print("Mean of salary is: ", salary['Experience Years'].mean())
print("Mean of Exp yrs is: ", salary['Salary'].mean())


X
```

```python
import matplotlib.pyplot as plt
plt.scatter(salary['Experience Years'], salary['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.show()
```

```python
plt.plot(salary['Experience Years'], salary['Salary'])
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)

# y = mx + c
c = model.intercept_
print("Intercept is: ", c)

m = model.coef_
print("Slope is: ", m)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
print("MAE: ", mean_absolute_error(y_test,y_pred))
print("MAE %error: ", mean_absolute_percentage_error(y_test,y_pred))
print("MSE: ", mean_squared_error(y_test,y_pred))
```

```python
import pandas as pd
predicted_salary = model.predict(pd.DataFrame([[2.5]], columns=['Experience Years']))
print(f"Predicted Salary for 2.5 years of experience: {predicted_salary[0]}")
```
